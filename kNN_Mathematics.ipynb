```python
def ed(data1, data2, length):
    
    distance = 0
    for x in range(length):
        distance += np.square(data1[x] - data2[x])
    return np.sqrt(distance)
```

```python
# Defining our KNN model
def knn(trainset, testset, k):
    distances = {}
    sort = {}
    
    length = testset.shape[1]
    print(testset)
    print(length)
    for x in range(len(trainset)):
        dist = ed(testset,trainset.iloc[x], length)
        #print(dist)
        distances[x] = dist[0]
       
    #print(distances)
    sorted_d = sorted(distances.items(), 
                      key=operator.itemgetter(1))
    #print(sorted_d)
    neighbors = [] 
    
    # Extracting top k neighbors
    for x in range(k):
        neighbors.append(sorted_d[x][0])
        #print(neighbors)
        #[[1,2,3,4,5],[3,4,5,6,7]]
    #### End of STEP 3.3
    classVotes = {}
    # Calculating the most freq class in the neighbors
    for x in range(len(neighbors)):
        responses = trainset.iloc[neighbors[x]]
        response = trainset.iloc[neighbors[x]][-1]
        #print(responses)
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
        print(classVotes)
    sortedVotes = sorted(classVotes.items(),
    key=operator.itemgetter(1), reverse=True)
    print(sortedVotes)
    return(sortedVotes[0][0])
```

```python
test=pd.DataFrame([[4.8,3,1.4,0.1]])
k = 5
result = knn(data, test, k)
print(result)
```
