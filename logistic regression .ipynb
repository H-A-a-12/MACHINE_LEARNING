```python
df.drop(['Gender','User ID'],axis=1)
```

```python
plt.scatter(df['Age'],df['EstimatedSalary'],marker='*',color='red')
```

```python
x=df.iloc[:,[2,3]].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x=StandardScaler()
x_scale=sc_x.fit_transform(x_train)
#y_scale=sc_x.fit_transform(y_train)
x_tests=sc_x.fit_transform(x_test)
```

```python
#Logistic Regression 
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_scale,y_train)
```

```python
classifier.score(x_tests,y_test)
```

```python
y=classifier.predict(x_tests)
#sc_x.inverse_transform(y)
y
```
